```python
faker_seed = 1
number_of_results = 125
```

```python
import datetime
from random import randint

fake = Faker()
# Change the seed below to get different results
Faker.seed(faker_seed)

def input_data(x):
   
    # pandas dataframe
    data = pd.DataFrame()
    for i in range(0, x):
        gender = np.random.choice(["M", "F"], p=[0.5, 0.5])
        state = fake.state_abbr(include_territories=False, include_freely_associated_states=False)

        data.loc[i,'ID']= randint(1, number_of_results*4)
        data.loc[i,'FIRST_NAME']= fake.first_name_male() if gender=="M" else fake.first_name_female()
        data.loc[i,'LAST_NAME']= fake.last_name()
        data.loc[i,'GENDER']= gender
        data.loc[i,'ADDRESS1']= fake.street_address()
        data.loc[i,'CITY']= fake.city()
        data.loc[i,'STATE_ABBR']= state
        data.loc[i,'POSTCODE']= fake.postalcode_in_state(state)
        data.loc[i,'COUNTRY_CODE']= fake.current_country_code()
        data.loc[i,'JOB_TITLE']= fake.job()
        data.loc[i,'PHONE_NUMBER']= str(fake.phone_number())
        data.loc[i,'EMAIL']= fake.email()
        data.loc[i,'SSN']= str(fake.ssn())
        data.loc[i,'BIRTH_DATE']= fake.date_between(datetime.date(1980, 1, 1), datetime.date(1995, 12, 31))
        data.loc[i,'HIRE_DATE']= fake.date_between(datetime.date(2015, 1, 1), 'today')
    return data
   

df_fake_data = input_data(number_of_results)
```

```python
df_fake_data.to_csv(f'/kaggle/working/faker_data_seed_{faker_seed}.csv', index=False)
```
